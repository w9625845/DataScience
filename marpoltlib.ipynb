```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(1,10)
y = np.arange(11,20)

plt.scatter(x,y,c='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2D graph')
plt.savefig('2d graph image')
```

```python
y= x*x
```

```python
plt.plot(x,y,'g*--')
plt.title('square graph')
plt.savefig('x-square graph')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
np.pi
x=np.arange(0,4*np.pi,0.5)
y=np.sin(x)
plt.title('sin graph')
plt.plot(x,y)
plt.savefig('sin graph')
```

```python
#bar plot
x= [4,5,6]
y=[6,8,9]
x2=[3,7,5]
y2=[7,6,9]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.savefig('bar graph')
```
